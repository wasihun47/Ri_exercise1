```python
import numpy as np
import matplotlib.pyplot as plt
from bandit_enviroment import Environment 
from bandit_reward import rewards_historys
from bandite_iteration import value_iterationsa
from bandit_polcyinterat import policy_iteration
from banditq import qlearnBandit
from bandit_ucb import UCB_Bandit

```

```python
# Example usage:
k = 10  # Number of arms
reward_distributions = np.random.normal(0, 1, k)  # Randomly generated reward distributions for each arm
environment = Environment(k, reward_distributions)
```

```python
# Run the bandit algorithm
bandit = rewards_historys(k)
rewards = bandit.run(environment, iterations=1000)
print(f" Reward: {rewards}")
# Plot the results
plt.plot(rewards)
plt.xlabel('Iterations')
plt.ylabel('Reward')
plt.title('Rewards Over Iterations')
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Environment:
    def __init__(self, k,  reward_distributions):
        self.k = k  # Number of arms
        self.reward_distributions = reward_distributions  # Reward distributions for each arm

    def get_reward(self, action):
        # Return a reward from the distribution corresponding to the selected action
        return np.random.normal(self.reward_distributions[action])

    def optimal_strategy(self):
        # Return the arm with the highest expected reward
        return np.argmax(self.reward_distributions)
```
